```python
from arcgis.features import FeatureLayer
api_key = input("api_key")
```

```python
from arcgis.gis import GIS
# Create a GIS object, as an anonymous user for this example
gis = GIS()
```

```python
lyr_url = f'https://geo.api.sbb.ch/site/rest/services/DGP_PUBLIC/netzelemente/FeatureServer/9?api_key={api_key}'
lyr = FeatureLayer(lyr_url)
lyr.properties
```

```python
import arcgis
import arcgis.map
map1 = gis.map()
map1: arcgis.map.Map
map1.content.add(lyr, options={'opacity':0.4})
map1
```
